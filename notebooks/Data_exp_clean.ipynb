```python
data = pd.read_csv('../data/measurements.csv')
```

```python
cols = [] 
for col in data.columns: 
    cols.append(col.lower().replace(' ', '_')) 
data.columns = cols

data.head()
```

## Droping the columns that have the highest % of NAs

```python
data = data.drop(['specials', 'refill_liters', 'refill_gas'], axis=1)
```

## Chosing a way of getting rid of the NA's from the colmun temp_inside

```python
data['temp_inside'].unique()
```

```python
data['temp_inside'] = pd.to_numeric(data['temp_inside'], errors='coerce')
data['temp_inside'] = data['temp_inside'].fillna(data['temp_inside'].mean())
```
